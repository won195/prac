```python
from glob import glob
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import koreanize_matplotlib
plt.style.use('ggplot')
```

```python
df_melt = df.melt(id_vars=['광역시도명', '시군구명', '날짜', '시간대별 시간', '강수 유형명', '습도값', '강수량 값', '기온값', '풍속값', '바람강도 유형명', '동쪽서쪽바람유형명', '풍향값', '풍향카테고리명'], var_name='음식', value_name='배달건수')
df_melt.head()
```

```python
sns.barplot(data=df_melt, y="음식", x="배달건수", errorbar=None)
```

```python
sns.catplot(data=df_melt, x='시간대별 시간', y='배달건수', errorbar=None, kind='bar', col='음식', col_wrap=4)
```

```python
px.histogram(df_melt, x='시간대별 시간', y='배달건수', histfunc='avg', text_auto=True, facet_col='음식', facet_col_wrap=4, title='시간대별 평균 음식 배달건수')
```

```python
sns.barplot(data=df_melt, x='시간대별 시간', y='배달건수', hue='강수 유형명', errorbar=None).set_title("강수 유형별 평균 배달건수")
```

```python
sns.catplot(data=df_melt, x='시간대별 시간', y='배달건수', hue='강수 유형명', kind='bar', col='음식', col_wrap=4, errorbar=None)
```
