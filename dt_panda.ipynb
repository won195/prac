# 연습

* 30대 여성 추출 -> 구매금액 높은순서로 정렬
* 파일로 저장

```python
cond1 = (df1['성별']=='여')
cond2 = (df1['연령대']=='30대이하')
df3 = df1[cond1 & cond2].sort_values(by='구매금액',ascending=False)
df3.shape
```

```python
df3.to_csv('result.csv',encoding='cp949')
```
